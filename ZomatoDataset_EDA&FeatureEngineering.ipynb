```python
#Reading country code data set which is in xlsx format
df_country=pd.read_excel('Zomato\Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
#combine two dataframes w.r.t country code
final_df=pd.merge(df, df_country, on='Country Code', how='left')
final_df.head(2)
#use concat w.r.t to this diff parameters you have to use it there
```

```python
## To Check Data Types
final_df.dtypes
```

```python
final_df.columns
```

```python
#How many dif countries are there and w.r.t those specific countries how many records are there
final_df.Country.value_counts()
```

```python
#get all the countries name w.r.t to specific records
country_names=final_df.Country.value_counts().index
country_names
```

```python
#get no. of records for that particular country name
number_of_records=final_df.Country.value_counts().values
number_of_records
```

```python
#creating a pie chart-Top 3 countries uses Zomato
plt.figure(figsize=(5,5))
plt.pie(number_of_records[0:3], labels=country_names[0:3], autopct='%1.2f%%')
plt.show()
```

Observation: 94.39% of zomato partnered restaurants are in India, 4.73% in US and 0.87% in UK
    
Zomato maximum records or transaction are from India after that USA abd then UK
             
             

```python
final_df.columns
```

```python
#features 'Aggregate rating', 'Rating color', 'Rating text'
#See w.r.t from which country more rating is actually coming 
#using group by, it will become dataframe group by object, then use size, we will see values
#convert it into dataframe, use reset_index? reseting the index of eachg row
#using rename, rename the column 0
#save this in a variable, varialbe will play imp role right now
ratings=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

```python
#find out this relationship(	Aggregate rating	Rating color	Rating text	Rating Count) w.r.t countries
#this is my dataset what type of visualizations I can draw from this, want to do some kind of EDA
```

Conclusion: Around 2148 raitngs have 0 rating

```python
#plot 
plt.figure(figsize=(7,7))
plt.pie(list(ratings['Rating Count']),labels=list(ratings['Rating color','Aggregate rating']))
plt.show()
```

## Observations imp*
1. When Rating is between 4.5 to 4.9 ---> Excellent
2. when Rating is between 4.0 to 4.4 ---> Very Good
3. when Rating is between 3.5 to 3.9 ---> Good
4. when Rating is between 3.0 to 3.4 ---> Average
5. when Rating is between 2.5 to 2.9 ---> Average
6. when Rating is between 1.8 to 2.4 ---> Poor
